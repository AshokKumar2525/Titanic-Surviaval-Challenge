```python
def fam_type(x):
    if x==1:
        return "alone"
    elif x<5:
        return "medium"
    else: return "large"
```

```python
train["family_size"] = train.family.apply(fam_type)
```

```python
test["family_size"] = test.family.apply(fam_type)
```
